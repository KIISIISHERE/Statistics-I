```python
import pandas as pd
import numpy as np
```

```python
#read and load the data
data=pd.read_csv("/content/Titanic Dataset.csv")
data.head()
```

```python
#check for null values
data.isnull().sum()
```

```python
#dropm null values
data.dropna(inplace=True)
```

```python
#mean age of passengers in The Titanic
mean_age=data['Age'].mean()
print(f"The Mean Age of Passengers in The Titanic is: {mean_age:.2f}")
```

```python
#meadian age of passengers in The Titanic
median_age=data['Age'].median()
print(f"The Median Age of Passengers in The Titanic is: {median_age:.2f}")
```

```python
#the most common age of the passengers in The Titanic
mode_age=data['Age'].mode()
print("The Most Common Age of Passengers in The Titanic is:", mode_age)
```
